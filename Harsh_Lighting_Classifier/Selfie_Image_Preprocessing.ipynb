```
def clean_targets(df):
  for column in df.columns:
    if df[column].dtype == 'int64':
      df[column] = df[column].apply(lambda x: 1 if x == 1 else 0)
  return df
```

```
targets = clean_targets(targets)
```

# Importing Images

```
def load_images(abs_path, folder, img_height, img_width, img_depth, 
                Normalize=False):
  img_folder = os.path.join(abs_path, folder)
  os.chdir(img_folder)

  X = []
  img_id = []

  for images in tqdm(os.listdir(img_folder)[:5):
    img_path = os.path.join(img_folder, images)
    img = image.load_img(img_path, target_size=(img_height, img_width, img_depth),
                         color_mode='rgb')
    img = image.img_to_array(img)
    if Normalize:
      img /= 255.0
    X.append(img)
    img_id.append(images)
  
  #X = np.array(X)
  
  df = pd.DataFrame()
  df['X_values'] = X
  df['img_id'] = img_id
  
  return df
```
